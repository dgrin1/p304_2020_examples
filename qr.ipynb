```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import qr
# unshift QR algorithm:
#using PyPlot
array = np.random.randint(10, size=(1000, 1000))
symm=np.matmul(array,array.T)
#A = rand(100,100); 
A=array
for k in range(1):
    (Q,R)=qr(A)
    A=R*Q

#spy(abs(A).>1e-4)
```
